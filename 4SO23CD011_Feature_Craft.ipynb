```python
transformed_data = data.pivot_table(index='State', columns='Season', values='Yield', aggfunc='mean')
print(transformed_data)
```

```python
transformed_data = data.pivot_table(index='Crop', columns='Crop_Year', values='Production', aggfunc='sum')
print(transformed_data)
```

```python
print("ACTUAL COLUMNS IN TRANSFORMED DATA:", transformed_data.columns)
long_df = pd.melt(transformed_data.reset_index(), id_vars='Crop', value_vars=[1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019], var_name='Crop_Year', value_name='Production')
print("\nAfter melting:")
print(long_df)
```

```python
average_yield_by_crop_year = data.groupby(['Crop', 'Crop_Year'])['Yield'].mean()
max_yield = average_yield_by_crop_year.max()
highest_yield_info = average_yield_by_crop_year[average_yield_by_crop_year == max_yield]
print("Crop and Year with the highest average yield:")
print(highest_yield_info)
```

```python
data['State'] = data['State'].str.lower().str.strip()
data['Dist'] = data['Dist'].str.lower().str.strip()

print("Unique values in 'State' column after cleaning:")
print(data['State'].unique())
print("\nUnique values in 'Dist' column after cleaning:")
print(data['Dist'].unique())
```

```python
average_yield_by_district_crop = data.groupby(['Dist', 'Crop'])['Yield'].transform('mean')
data['Production'].fillna(data['Area '] * average_yield_by_district_crop, inplace=True)
print("Missing values in 'Production' after imputation:", data['Production'].isnull().sum())
```

```python
odisha_wheat_production = data[(data['State'] == 'odisha') & (data['Crop'] == 'Wheat')]['Production'].sum()
print("Total wheat production in Odisha:", odisha_wheat_production)
```

```python
data['Calculated_Yield'] = data['Production'] / data['Area ']
data['Yield_Difference'] = data['Calculated_Yield'] - data['Yield']
sum_of_yield_differences = data['Yield_Difference'].sum()
print("Sum of the differences between Calculated_Yield and original Yield:", sum_of_yield_differences)
```

```python
negative_values = data[(data['Area '] < 0) | (data['Production'] < 0) | (data['Yield'] < 0)]
print("Rows with negative Area, Production, or Yield values:", len(negative_values))
production_positive_area_zero = data[(data['Production'] > 0) & (data['Area '] == 0)]
print("Rows with Production > 0 and Area == 0:", len(production_positive_area_zero))
future_years = data[data['Crop_Year'] > 2024]
print("Rows with Crop_Year > 2024:", len(future_years))
```
